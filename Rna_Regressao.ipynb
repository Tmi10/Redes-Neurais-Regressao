DESENVOLVER UM ALGORITMO DE REDES NEURAIS ARTFICIAIS DE REGRESSÃO : 

- De acordo com os resultados abaixo o algoritmo com o escalonamento dos previsores foi o que obteve o melhor desempenho com 86.77 % e o sem fazer o escalonamento deu 75.71 %

Sendo assim o algoritmo de RNA de Regressão com previsores escalonado foi o que obteve o melhor desempenho

```python
import pandas as pd
import numpy as np 
```

```python
df = pd.read_csv('insurance.csv', sep=',' , encoding='iso-8859-1')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
from sklearn.preprocessing import LabelEncoder

df.iloc[:,1] = LabelEncoder().fit_transform(df.iloc[:,1])
df.iloc[:,4] = LabelEncoder().fit_transform(df.iloc[:,4])
df.iloc[:,5] = LabelEncoder().fit_transform(df.iloc[:,5])
```

```python
df.head(20)
```

```python
independente = df.iloc[:, 0:6].values
independente
```

```python
dependente = df.iloc[:, 6:7].values
dependente
```

```python
from sklearn.model_selection import train_test_split
x_treino, x_teste, y_treino, y_teste = train_test_split(independente, dependente, test_size = 0.3, random_state = 0)
```

```python
x_teste.shape
```

```python
y_teste.shape
```

```python
x_treino.shape
```

```python
y_treino.shape
```

```python
from sklearn.neural_network import MLPRegressor

redes = MLPRegressor(hidden_layer_sizes=(100, 100), activation='relu', verbose=True, max_iter=2000,
                    solver='adam', random_state = 12)
```

```python
redes.fit(x_treino, y_treino)
```

```python
redes.n_layers_
```

```python
redes.score(x_treino, y_treino)
```

```python
# Teste

redes.score(x_teste, y_teste)
```

```python
previsoes_teste_rna = redes.predict(x_teste)
```

```python
# Métricas

from sklearn.metrics import mean_absolute_error, mean_squared_error

# Erro médio Absoluto
mean_absolute_error(y_teste, previsoes_teste_rna)
```

```python
# Raiz do erro quadrático médio (RMSE)
np.sqrt(mean_squared_error(y_teste, previsoes_teste_rna))
```

```python
# Validação Cruzada 

from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
```

```python
# Separando os dados em folds

kfold = KFold(n_splits = 12, shuffle=True, random_state = 5)
```

```python
# Criando o modelo
from sklearn.neural_network import MLPRegressor
modelo_rna = MLPRegressor(hidden_layer_sizes=(100, 100), activation='relu', verbose=True, max_iter=2000,
                    solver='adam', random_state = 12)
resultado_rna = cross_val_score(modelo_rna, independente, dependente, cv = kfold)
```

```python
resultado_rna
```

```python
# Usamos a média e o desvio padrão
print("Acurácia Média: %.2f%%" % (resultado_rna.mean() * 100.0))
```

```python
# Escalonamento 

from sklearn.preprocessing import StandardScaler
x_scaler = StandardScaler()
x_treino_scaler = x_scaler.fit_transform(x_treino)
```

```python
# Teste 

redes.score(x_teste_scaler, y_teste_scaler)
```
